```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
training_set = pd.read_csv(r"D:\MachineLearningLibraries\Linear_regression\traning_set.csv")

x_train = training_set["YearsExperience"].values
y_train = training_set["Salary"].values
```

```python
plt.scatter(x_train, y_train, color = 'blue')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
```

```python
def cost_function(x, y, m, b):
    n = len(x)
    cost_sum = 0

    for i in range(n):
        f = m * x[i] + b
        cost = (f- y[i])**2
        cost_sum += cost

    total_cost = (1/(2*n)) * cost_sum
    return total_cost
```

```python
def gradient_function(x, y, m, b):
    n = len(x)
    m_gradient = 0
    b_gradient = 0

    for i in range(n):
        f = m * x[i]+b

        m_gradient += (f - y[i]) * x[i]
        b_gradient += (f-y[i])
    
    m_gradient = (1/n) * m_gradient
    b_gradient = (1/n) * b_gradient

    return m_gradient, b_gradient
```

```python
def gradient_decent(x, y, alpha, iterations):
    m = 0
    b = 0

    for i in range(iterations):
        m_gradient, b_gradient = gradient_function(x, y, m, b)

        m = m-alpha*m_gradient
        b = b - alpha* b_gradient

        print(f"iteration{i}: cost {cost_function(x, y, m, b)}")

    return m, b


    
```

```python
learning_rate = 0.01
iterations = 10000

final_m, final_b = gradient_decent(x_train, y_train,learning_rate, iterations,)
print(f"m: {final_m:.4f}, b: {final_b:.4f}")
```

```python
plt.scatter(x_train, y_train, label = 'Data Points', color = 'blue')
x_vals = np.linspace(min(x_train), max(x_train), 100)
y_vals = final_m * x_vals + final_b
plt.plot(x_vals, y_vals, color = 'red', label = 'Regression Line')

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression: Experience vs Salary")
plt.legend()
plt.show()
```
